```python
import pandas as pd

df= pd.read_csv('/content/drug-use-by-age.csv')

df.head()

print(df.info())
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=10, kde=True, color='Green')

plt.xlabel("Age")
plt.ylabel("frequency")
plt.title("Age Distribution of Drug user")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
